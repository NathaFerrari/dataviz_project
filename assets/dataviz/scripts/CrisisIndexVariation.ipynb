```python
import pandas as pd
```

```python
df = pd.read_csv("../DataSets/pivot_wellbeing_index.csv")
df = df.set_index("Reference area")

crisis = {
    "crisis_08_10" : (2007, 2009),
    "covid" : (2019, 2021),
    "ukrenian_war" : (2021, 2022)
}
```

```python
changes = {}
for name, (start_year, end_year) in crisis.items():
    col_start = str(start_year)
    col_end = str(end_year)
    # Make sure both years exist in the columns
    if col_start in df.columns and col_end in df.columns:
        changes[name] = df[col_end] - df[col_start]

# Build a DataFrame with the changes
changes_df = pd.DataFrame(changes)
changes_df.to_csv("../DataSets/wellbeing_index_changes.csv")
```
